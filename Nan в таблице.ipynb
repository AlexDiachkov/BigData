# удаление столбцов с большим количеством пропусков 
## (более 90%)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Diamond Price Prediction.csv')
df.head(10)
```

```python
df['X'] = np.nan
df
```

```python
df.info()
```

```python
many_null_colms = [col for col in df.columns if df[col].isnull().sum() / df.shape[0] > 0.9]
df = df.drop(many_null_colms, axis=1)
```

```python
df
```

## кодировка пропусков

```python
df['Y'] = np.nan
df
```

```python
df['Y'].value_counts(dropna=False)
```

```python
df['Y'] = df['Y'].fillna(-9999)
```

```python
df
```

# замена пропусков медианой всех значений

```python
df["Y"] = df["Y"].fillna(df['Y'].median())
```
